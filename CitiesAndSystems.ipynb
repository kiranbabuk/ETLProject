```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file="Resources/cities.csv"
cities_df=pd.read_csv(csv_file)
cities_df.head()
```

```python
cities_df[['lat','long']] = cities_df.coords.str.split(" ",expand=True,)
cities_df['lat'] = cities_df['lat'].map(lambda x: x.lstrip('POINT('))
cities_df['long'] = cities_df['long'].map(lambda x: x.rstrip(')'))
cities_df = cities_df[["id","name","start_year","country","country_state","lat","long"]].copy()
cities_df.head()
```

```python
csv_file="Resources/systems.csv"
systems_df=pd.read_csv(csv_file)
systems_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETLProject"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
cities_df.to_sql(name='newcities', con=engine, if_exists='append', index=False)
systems_df.to_sql(name='systems', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from newcities', con=engine).head()
```

```python
pd.read_sql_query('select * from systems', con=engine).head()
```
