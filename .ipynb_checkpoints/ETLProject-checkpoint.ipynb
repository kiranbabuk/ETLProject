```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETLProject"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
stations_data_df = pd.read_csv("Resources/stations.csv")
lines_df = pd.read_csv("Resources/lines.csv")
track_lines_df = pd.read_csv("Resources/track_lines.csv")
cities_df = pd.read_csv("Resources/cities.csv")
systems_df = pd.read_csv("Resources/systems.csv")
tracks_data_df = pd.read_csv("Resources/tracks.csv")
stationlines_df = pd.read_csv("Resources/station_lines.csv")
```

```python
stations_data_df[['Lat','Lang']] = stations_data_df.geometry.str.split(" ",expand=True,)
stations_data_df['Lat'] = stations_data_df['Lat'].map(lambda x: x.lstrip('POINT('))
stations_data_df['Lang'] = stations_data_df['Lang'].map(lambda x: x.rstrip(')'))
stations_data_df.head()
```

```python
cities_df[['Lat','Lang']] = cities_df.coords.str.split(" ",expand=True,)
cities_df['Lat'] = cities_df['Lat'].map(lambda x: x.lstrip('POINT('))
cities_df['Lang'] = cities_df['Lang'].map(lambda x: x.rstrip(')'))
cities_df.head()
```

```python
tracks_data_df = tracks_data_df[['id', 'closure', 'length', 'city_id']].copy()
tracks_data_df.head()
```

```python
stations_data_df.to_sql(name='newstations', con=engine, if_exists='append', index=False)
lines_df.to_sql(name='lines', con=engine, if_exists='append', index=False)
track_lines_df.to_sql(name='tracklines', con=engine, if_exists='append', index=False)
cities_df.to_sql(name='newcities', con=engine, if_exists='append', index=False)
systems_df.to_sql(name='systems', con=engine, if_exists='append', index=False)
stationlines_df.to_sql(name='stationlines', con=engine, if_exists='append', index=False)
tracks_data_df.to_sql(name='newtracks', con=engine, if_exists='append', index=False)
```
