```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
lines_df = pd.read_csv("Resources/lines.csv")
track_lines_df = pd.read_csv("Resources/track_lines.csv")
```

```python
lines_df.head()
```

```python
track_lines_df.head()
```

```python
track_lines_df.describe()
```

```python
#Connect to the database
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETLProject"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
lines_df.to_sql(name='lines', con=engine, if_exists='append', index=False)
track_lines_df.to_sql(name='track_lines', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from lines', con=engine).head()
```

```python
pd.read_sql_query('select * from track_lines', con=engine).head()
```
